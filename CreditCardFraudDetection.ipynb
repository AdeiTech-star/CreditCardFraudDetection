## Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#load the dataset to our dataframe
data = pd.read_csv('creditcard.csv')
```

```python
#print first 5 rows of the data set
data.head()
# V1 means trasaction number
#Time is the time lapse of when the transaction happened from the initial time transaction
#Class is what we will be predicting: 0 and 1 -> not fraud / fraud
```

```python
#cehck if we have any missing values in each column
data.isnull().sum()
#this is to also confirm that we do not have any missing values aswell
```

```python
# check the distribution of the the class: either being fraudulent or legit - > this helps in fairness when we'll be checking the accuracy
data['Class'].value_counts()
# we can see that the data is unbalanced. 
```

## This dataset is highly imbalanced and we will likely get incorrect predictions.
### 0 --> legit transation
### 1 --> fraudulent transaction

# We will now be handling imbalanced data

```python
# separate data for analysis:
legit = data[data.Class == 0]
fraud = data[data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# now lets get the statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transations
data.groupby('Class').mean()
```

##### For us to deal with this unbalanced data, we will be using 
##### 1) Undersampling: so undersampling is a technique we use to balance uneven dataset by keeping all the data in the minority and decreasing the size of the majority                                      

```python
legit_data = legit.sample(n=492)
```

```python
#concatinate two df
new_data = pd.concat([legit_data, fraud], axis=0) # now we have created a new dataset. axis =0 are rows
```

```python
new_data.head()
```

```python
new_data['Class'].value_counts()
#Now our data is balanced
```

```python
new_data.groupby('Class').mean()
```

### Test Train Split

```python
X = new_data.drop('Class', axis = 1) # we drop the Class column because this is our feature and we'll be predictiong it
Y = new_data['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=42) # 20% goes to the test, stratify ensures that the training and testing sets maintain the same proportion of samples for each class as in the original dataset
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

#### Model training: Logistic Regression because this is a binary classification problem

```python
model = LogisticRegression(max_iter=1000)
```

```python
# now we train our model
model.fit(X_train,y_train)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
```

```python
model = make_pipeline(StandardScaler(), LogisticRegression(max_iter=1000))
```

```python
model.fit(X_train, y_train)
```

#### Evaluate the model

#### Accuracy score

```python
X_train_pred = model.predict(X_train)
training_data_acc = accuracy_score(X_train_pred, y_train)

```

```python
print('Accuracy on training data: ', training_data_acc)
```

```python
#Accuracy on testing data
x_test_pred = model.predict(X_test)
test_data_accuracy = accuracy_score(x_test_pred, y_test)
```

```python
print('Accuracy on test data: ', test_data_accuracy)
```
